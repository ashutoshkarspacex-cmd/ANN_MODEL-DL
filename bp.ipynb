```python
import numpy as np

# ==========================================
# 1. Activation Functions & Derivatives
# ==========================================
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def sigmoid_derivative(z):
    s = sigmoid(z)
    return s * (1 - s)

def relu(z):
    return np.maximum(0, z)

def relu_derivative(z):
    return (z > 0).astype(float) # Returns 1 if z > 0, else 0

# ==========================================
# 2. Initialize Network Parameters
# ==========================================
# Input space: 2 features. Hidden space: 64 neurons.
input_size = 2
hidden_size = 64
output_size = 1

# Randomly initialize weights and biases
# W1 dimensions: (64, 2)
W1 = np.random.randn(hidden_size, input_size) * 0.01 
b1 = np.zeros((hidden_size, 1))

# W2 dimensions: (1, 64)
W2 = np.random.randn(output_size, hidden_size) * 0.01 
b2 = np.zeros((output_size, 1))

# Dummy Training Data (1 example)
X = np.array([[0.5], [0.8]]) # Dimensions: (2, 1)
Y = np.array([[1]])          # True label: 1

# ==========================================
# 3. The Forward Pass
# ==========================================
# Hidden Layer (ReLU)
Z1 = np.dot(W1, X) + b1      # Raw logits for hidden layer. Dim: (64, 1)
A1 = relu(Z1)                # Activations for hidden layer. Dim: (64, 1)

# Output Layer (Sigmoid)
Z2 = np.dot(W2, A1) + b2     # Raw logit for output layer. Dim: (1, 1)
A2 = sigmoid(Z2)             # Final probability prediction. Dim: (1, 1)

# ==========================================
# 4. BACKPROPAGATION (The Math We Discussed!)
# ==========================================

# First, calculate the derivative of the Binary Cross-Entropy Loss
# dA2 formula: - (Y/A2) + (1-Y)/(1-A2)
dA2 = -(np.divide(Y, A2) - np.divide(1 - Y, 1 - A2)) # Dim: (1, 1)

# -- STEP 1: Output Layer Logit Error --
# Multiply loss error by Sigmoid derivative
dZ2 = dA2 * sigmoid_derivative(Z2) 
# Note: In practice, BCE + Sigmoid mathematically simplifies nicely to just: dZ2 = A2 - Y

# -- STEP 2: Update Final Weights --
# Multiply logit error by the hidden layer activations (transposed)
dW2 = np.dot(dZ2, A1.T)      # Dim: (1, 1) dot (1, 64) -> (1, 64)
db2 = np.sum(dZ2, axis=1, keepdims=True)

# -- STEP 3: Pass Blame to Hidden Layer --
# Multiply logit error by transposed weights
dA1 = np.dot(W2.T, dZ2)      # Dim: (64, 1) dot (1, 1) -> (64, 1)

# -- STEP 4: Hidden Layer Logit Error --
# Pass the error backward through the ReLU gate
dZ1 = dA1 * relu_derivative(Z1) # Dim: (64, 1) * (64, 1) -> (64, 1)

# -- STEP 5: Update Initial Weights --
# Multiply hidden logit error by the raw Input data
dW1 = np.dot(dZ1, X.T)       # Dim: (64, 1) dot (1, 2) -> (64, 2)
db1 = np.sum(dZ1, axis=1, keepdims=True)

# ==========================================
# 5. The Optimizer (Vanilla Gradient Descent)
# ==========================================
learning_rate = 0.01

# Update the physical weights by subtracting the calculated gradients
W1 = W1 - learning_rate * dW1
b1 = b1 - learning_rate * db1
W2 = W2 - learning_rate * dW2
b2 = b2 - learning_rate * db2

print("Forward Pass Prediction (A2):", A2[0][0])
print("Shape of dW1 perfectly matches W1:", dW1.shape == W1.shape)
print("Shape of dW2 perfectly matches W2:", dW2.shape == W2.shape)
```
